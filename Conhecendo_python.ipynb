```python
#Sistema de conferecia de faturas

from datetime import datetime

nome = input('digite seu nome :')
idade = int(input('Digite a idade :'))
endereço = input('Digite o seu endereço')
profissao = input('Digite a sua profissão:')
salario = float(input('Digite o seu salario:'))
valor_fatura_do_cartão = float(input('Digite aqui o valor da sua fatura do cartão de credito'))
vencimento_fatura = (input('Digite a data do vencimento da fatura, (dd/mm/yyyy)'))
vencimento_ajustado = datetime.strptime(vencimento_fatura, '%d/%m/%Y')
```

```python
#Condição do sistema

diferença = salario - valor_fatura_do_cartão
data_da_validação = datetime.now()


#Verifica se o salário cobre a fatura

if salario < valor_fatura_do_cartão:
    print(f"Falta R$ {diferença:.2f} para pagar a fatura.")

    
    # Pergunta sobre depósito para completar o pagamento
   
    while True:
        
        novo_deposito = input('Deseja fazer um depósito? (Sim/Não): ').strip().lower()

        if novo_deposito in ['sim', 's']:
            Valor_deposito = float(input('Digite o valor a ser depositado: R$ '))
            salario += Valor_deposito
            print(f"Depósito realizado! Novo saldo: R$ {salario:.2f}")
            break

        elif novo_deposito in ['não', 'nao', 'n']:
            print(f"Depósito não selecionado. Ainda falta R$ {diferença:.2f}.")
            break

        else:
            print('Resposta inválida, digite somente "Sim" ou "Não".')


if salario >= valor_fatura_do_cartão:
    print("O salário cobre a fatura.")

 
    #  Verifica vencimento
    
    if data_da_validação > vencimento_ajustado:
        print('Fatura vencida, solicite a reemissão.')
    else:
        print('Fatura Paga com sucesso! ✔️')

else:
    print(f"Ainda não cobre. Falta R$ {valor_fatura_do_cartão - salario:.2f}")
    print("Fatura não paga.")
```
