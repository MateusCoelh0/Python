```python
# Resumo de funçoes e metodos estatísticos

df['Idade'].max() # Maior idade
df['Idade'].min() # Menor idade
df['Idade'].mean() # Média de idade
df['Idade'].std() # Desvio padrão da idade
df['Idade'].sum() # Soma total das idades
df['Idade'].count() # Contagem de registros na coluna idade
df['Idade'].median() # Mediana da idade
df['Idade'].mode() # Moda da idade
df['Idade'].var() # Variância da idade
df['Idade'].describe() # Resumo estatístico da coluna idade
df['Gasto_Total'].describe() # Resumo estatístico da coluna Gasto_Total
df.describe() # Resumo estatístico de todas as colunas numéricas do DataFrame
df.info() # Informações gerais sobre o DataFrame, incluindo tipos de dados e contagem de valores não nulos
df.head() # Exibe as primeiras 5 linhas do DataFrame
df.tail() # Exibe as últimas 5 linhas do DataFrame
df.sample(10) # Exibe 10 linhas aleatórias do DataFrame
df.nunique() # Número de valores únicos em cada coluna do DataFrame
df.corr() # Matriz de correlação entre as colunas numéricas do DataFrame
df.count() # Contagem de valores não nulos em cada coluna do DataFrame
```

```python
# Atributos do DataFrame

df.columns # Lista os nomes das colunas do DataFrame
df.index # Mostra o índice (linhas) do DataFrame
df.dtypes # Mostra os tipos de dados de cada coluna do DataFrame
df.shape # Mostra a forma (número de linhas e colunas) do DataFrame
df.size # Mostra o número total de elementos no DataFrame (linhas * colunas)
df.values # Retorna os valores do DataFrame como um array NumPy
df.empty # Verifica se o DataFrame está vazio (True ou False)
```

EXERCICIO 3 FUNCOES Objetivo:

Utilize o DataFrame fornecido para agrupar os clientes por Estado, somar os gastos totais de cada grupo, e exibir apenas os grupos cujo
total ultrapasse R$ 5.000,00.

```python
# Exercicio de agrupamento com filtragem
# Filtragem de grupos após o agrupamento

df_agrupados = df.groupby('Estado')['Gasto_Total'].sum() 
df_filtro_por_total = df_agrupados[df_agrupados > 5000]  
print(df_filtro_por_total)
```
